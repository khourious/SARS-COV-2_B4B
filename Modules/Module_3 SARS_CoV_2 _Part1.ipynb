# Pangolim

PANGOLIN é um acrônimo para Phylogenetic Assignment of Named Global Outbreak Lineages. É um software que atribui seu genoma SARS-CoV-2 à linhagem SARS-CoV-2 relacionada mais próxima no contexto global com base nas mutações na sequência de consulta. O Pangolin pode ser acessado em seu aplicativo da Web, que permite fazer upload de seus arquivos FASTA aqui: https://pangolin.cog-uk.io/.

## Instalando o Pangolin para detecção de variantes
O Pangolin pode ser instalado dentro do conda executando o código abaixo:

````python
!conda install -c bioconda -c conda-forge -c defaults pangolin=4.1.1
````

````python
# Atualize a dependência tabular para evitar erros associados ao Pangolin
!pip install tabulate==0.8.10
````

Para verificar se o Pangolin foi instalado corretamente, execute o seguinte comando:

````python
!pangolin --help
````

## Iniciando o Pangolin no arquivo multi-FASTA
Agora você instalou o Pangolin com sucesso e exibiu as opções de ajuda que mostram os comandos e opções que você pode ajustar para executar sua análise com o software.

Vamos agora tentar identificar as variantes dos genomas no arquivo FASTA mesclado usando o Pangolin com as configurações padrão; e direcionar nossa saída para a pasta `resultados`. Por padrão, o arquivo de saída será chamado `lineage_report.csv`.

````python
!pangolin --outdir results combined_genomes.fasta
````

**NOTA**: Baixe o arquivo `lineage_report.csv` para o seu computador e abra-o com um software de planilha como o MS Excel.

# Nextclade

Nextclade é outro software fantástico que pode identificar variantes em seus genomas como o Pangolin. No entanto, ele possui recursos adicionais em seu relatório, como o número de mutações no genoma e onde elas estão localizadas. Nextclade pode ser acessado no aplicativo da web: https://clades.nextstrain.org/.

## Instalando o Nextclade

Instale o Nextclade com conda usando o seguinte comando e teste a instalação exibindo a página de ajuda.

````python
!conda install -c bioconda nextclade
````

````python
# Para verificar se o Nextclade foi instalado corretamente
!nextclade --help
````

## Executando NextClade

Antes de executar o Nextclade nos genomas montados, primeiro baixe o último conjunto de dados Nextclade disponível para SARS-CoV-2.

````python
# Baixe a pasta do conjunto de dados SARS-CoV-2 para NextClade
!nextclade dataset get --name sars-cov-2 --output-dir sars-cov-2-dataset
````

Em seguida, identifique os clados dos genomas no arquivo multi-FASTA.

````python
# Execute o NextClade em seus arquivos FASTA concatenados para gerar o arquivo de saída TSV
!nextclade run --output-tsv nextclade_report.tsv --input-dataset sars-cov-2-dataset combined_genomes.fasta
````

**NOTA**: Baixe o arquivo "nextclade_report.tsv" e abra-o com um software de planilha em seu computador, como o MS Excel.
